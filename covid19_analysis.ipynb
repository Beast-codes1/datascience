```python
#CASE STUDY: COVID-19 DATA ANALYSIS

#import the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the dataset
dataset = pd.read_csv("Downloads/data_science/data science/covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
#delete the useless columns
df = dataset.drop(["Lat","Long"], axis = 1,inplace = True)
```

```python
#aggrigate teh rows by the country
corona_dataset_aggregated = dataset.groupby("Country/Region").sum()
```

```python
#maximum infection rate

corona_dataset_aggregated.loc["China"].diff().max()   
```

```python
corona_dataset_aggregated.loc["India"].diff().max()  
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()   
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []

for c in countries:
    numeric_data = pd.to_numeric(corona_dataset_aggregated.loc[c], errors='coerce')
    max_infection_rate = numeric_data.diff().max()
    max_infection_rates.append(max_infection_rate)
   # max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["Max_infection_rates"] = max_infection_rates
```

```python
corona_dataset_aggregated
```

```python
#create a new dataframe
corona_data = pd.DataFrame(corona_dataset_aggregated["Max_infection_rates"])
```
